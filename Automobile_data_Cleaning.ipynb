```python
df1 = df.replace(['?','','Null'],np.nan)
df1.head()
```

```python
df["normalized-losses"] = pd.to_numeric(df["normalized-losses"], errors="coerce")
avg = df["normalized-losses"].mean()
df["normalized-losses"]=df["normalized-losses"].fillna(avg, inplace=True)
```

```python
df["bore"] = pd.to_numeric(df["bore"], errors="coerce")
avg = df["bore"].mean()
df["bore"]=df["bore"].fillna(avg, inplace=True)
```

```python
df["stroke"] = pd.to_numeric(df["stroke"], errors="coerce")
avg = df["stroke"].mean()
df["stroke"]=df["stroke"].fillna(avg, inplace=True)
```

```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
avg = df["horsepower"].mean()
df["horsepower"]=df["horsepower"].fillna(avg, inplace=True)
```

```python
df["peak-rpm"] = pd.to_numeric(df["peak-rpm"], errors="coerce")
avg = df["peak-rpm"].mean()
df["peak-rpm"]=df["peak-rpm"].fillna(avg, inplace=True)
```

```python
mode = df['num-of-doors'].mode()[0]
df['num-of-doors']=df['num-of-doors'].fillna(mode)
```
